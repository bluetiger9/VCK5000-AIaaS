### Imports

```python
import requests
import json
from IPython.display import Image
```

### API Server

```python
#apiServer = "http://localhost:8080/api/v1"
apiServer = "http://x570-test.local:8080/api/v1"
```

### Input Image

```python
image = "sample_classification.jpg"
Image(filename=image)
```

### Classify Request (remote)

```python
files = {'image': (image, open(image, 'rb') ) }         
model = "resnet50"

# make request
response = requests.post(apiServer + '/images/classify?model=' + model, files=files)
response.status_code
```

### Inspect results:

```python
jsonResponse = json.loads(response.content)

for result in jsonResponse['data']['results']:
    print("Class: {} Score: {}".format(result['class'], result['score']))
```
