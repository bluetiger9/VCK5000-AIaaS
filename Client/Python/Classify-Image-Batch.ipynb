### Imports

```python
import requests
import json
from IPython.display import Image
```

### API Server

```python
#apiServer = "http://localhost:8080/api/v1"
apiServer = "http://x570-test.local:8080/api/v1"
```

### Input Image

```python
images = [
    "sample_classification_1.jpg",
    "sample_classification_2.jpg",
    "sample_classification_3.jpg",
    "sample_classification_4.jpg",
    "sample_classification_5.jpg",
    "sample_classification_6.jpg",
    "sample_classification_7.jpg",
    "sample_classification_8.jpg",
]

for image in images:
    display(Image(filename=image))
```

### Classify Request (batch / remote)

```python
files = []
for image in images:
    files.append(('images', open(image, 'rb')))

model = "resnet50"

# make request
response = requests.post(apiServer + '/images/classify/batch?model=' + model, files=files)
response.status_code
```

### Inspect results:

```python
jsonResponse = json.loads(response.content)

for result in jsonResponse['data']['results']:
    print("Image: {}".format(result['image']))
    for result2 in result['results']:
        print("Class: {} Score: {}".format(result2['class'], result2['score']))
    print()
```
